# yes24 크롤링

```python
def parse_yes24(isbn):
    
    session = requests.session()
    session.proxies = {}
    session.proxies['http'] = 'socks5h://localhost:9050'
    session.proxies['https'] = 'socks5h://localhost:9050'
    
    book = {'isbn' : isbn,
            '제목' : '없음',
            '저자' : '없음',
            '역자' : '없음',
            '별점' : '없음',
            '판매지수' : '없음',
            '리뷰' : '없음',
            '카테고리' : '없음',
            '책소개' : '없음',
            '저자소개' : '없음',
            '추천사' : '없음',
            '목차' : '없음',
            '책속으로' : '없음',
            '출판사리뷰' : '없음',
            '줄거리' : '없음'
           }
    
    search_addr = 'http://www.yes24.com/searchcorner/Search?keywordAd=&keyword=&domain=ALL&qdomain=%C0%FC%C3%BC&query='
    search_page = session.get(search_addr + isbn)
    goodlist_soup = BeautifulSoup(search_page.text, 'html.parser')

    try:
        goodlist_el = goodlist_soup.select_one('.goodsList_list')
        goods_title_list = goodlist_el.select('.goods_icon')
        goods_title = goods_title_list[0].select_one('strong').text
        goods_link = goods_title_list[0].select_one('a[href]')['href']
    except:
        return book

    # 상품 상세 페이지 - 
    base_url = 'http://www.yes24.com'
    rr = session.get(base_url + goods_link)
    
    
    
#     btns = driver.find_elements_by_class_name('txt_open')
#     for btn in btns:
#         try:
#             btn.click()
#         except:
#             pass
#     item_page = driver.page_source
    item_soup = BeautifulSoup(rr.text, 'html.parser')

    # 저자, 역자, 책 제목
    try:
        author_line = item_soup.select_one('meta[name="author"]')['content']
        title = item_soup.select_one('meta[property="og:title"]')['content']
        auth_trans = author_line.split('/')
        author = '없음'
        translator = '없음'
        for a in auth_trans:
            if '저' in a:
                author = a.replace(' 저', '').strip()
            if '역' in a:
                translator = a.replace(' 역', '').strip()
    except:
        author = '없음'
        translator = '없음'
        title = '없음'
    book['저자'] = author
    book['역자'] = translator
    book['제목'] = title

    # 별점, 판매지수, 리뷰
    rating_elements = item_soup.select_one('.gd_ratingArea')
    try:
        rating_score = rating_elements.em.text
    except:
        rating_score = '없음'
    book['별점'] = rating_score
    
    try:
        sales_element = rating_elements.select_one('.gd_sellNum')
        sales_point = re.findall('\d+', sales_element.text)[0]
    except:
        sales_point = '없음'
    book['판매지수'] = sales_point
        
    try:
        review_count_el = rating_elements.select_one('.gd_reviewCount')
        review_count = re.search('\d+', review_count_el.text).group()
    except:
        review_count = '없음'
    book['리뷰'] = review_count

    # 카테고리
    try:
        category_element = item_soup.select_one('.yesAlertDl')
        category = category_element.text.replace('\n', '').replace('\xa0', '').replace('카테고리 분류', '').replace('>', ' > ').strip()
    except:
        category = '없음'
    book['카테고리'] = category

    # 책 소개
    try:
        book_intro_element = item_soup.select_one('#infoset_introduce')
        book_intro_s = book_intro_element.select_one('.infoWrap_txtInner').textarea.text
        book_intro = re.sub('\<[^>]*\>', '', book_intro_s).strip()
    except:
        book_intro = '없음'
    book['책소개'] = book_intro

    # 저자 소개
    try:
        author_intro_element = item_soup.select_one('#contents_authoretc_text1')
        author_intro = author_intro_element.textarea.text.replace('<br/>', '\n').strip()
    except:
        author_intro = '없음'
    book['저자소개'] = author_intro
    
    # 추천사
    try:
        recommend_el = item_soup.select_one('div#infoset_nomiCmt')
        recommend = recommend_el.find('div', {'class' : 'infoWrap_txt'}).text.strip()
    except:
        recommend = '없음'
    book['추천사'] = recommend
    
    # 목차
    try:
        table_el = item_soup.select_one('div#infoset_toc')
        table = table_el.textarea.text.replace('<br/>', '\n')
    except:
        table = '없음'
    book['목차'] = table
    
    # 책속으로
    try:
        inbook_el = item_soup.select_one('div#infoset_inBook').select_one('.txtContentText')
        inbook = inbook_el.text
    except:
        inbook = '없음'
    book['책속으로'] = inbook

    # 출판사 리뷰
    try:
        pub_intro_element = item_soup.select_one('#infoset_pubReivew')
        pub_intro = pub_intro_element.find('div', {'class' : 'infoWrap_txt'}).text.strip()
    except:
        pub_intro = '없음'
    book['출판사리뷰'] = pub_intro
    
    # 줄거리
    try:
        summary_el = item_soup.select_one('#infoset_summary')
        summary = summary_el.find('div', {'class' : 'infoWrap_txt'}).text.strip()
    except:
        summary = '없음'
    book['줄거리'] = summary
        
    return book
```

# 목록 불러오기

```python
book_specs = pd.read_csv('./new_isbns_1804_2103.csv', low_memory=False)
isbns_from_csv = book_specs.loc[:, 'ISBN'].tolist()
isbns = [str(x).strip() for x in isbns_from_csv]

len(isbns)
```

# 대량 크롤링

```python
num = 140000
count_save = num - 10000
results = []
for isbn in tqdm(isbns[num - 10000:num]):
    time.sleep(1)
    try:
        result = parse_yes24(isbn)
        results.append(result)
    except:
        result = {'isbn' : isbn,
                  '제목' : '없음',
                  '저자' : '없음',
                  '역자' : '없음',
                  '별점' : '없음',
                  '판매지수' : '없음',
                  '리뷰' : '없음',
                  '카테고리' : '없음',
                  '책소개' : '없음',
                  '저자소개' : '없음',
                  '추천사' : '없음',
                  '목차' : '없음',
                  '책속으로' : '없음',
                  '출판사리뷰' : '없음',
                  '줄거리' : '없음'
                 }
        results.append(result)

    count_save += 1
    
    if count_save % 1000 == 0:
        file_name = f'yes_{count_save}_newisbn.csv'
        results_df = pd.DataFrame(results)
        results_df.to_csv('./crawled_yes/' + file_name, encoding='utf-8-sig')
        results = []
        print(f'saving {count_save}th results to {file_name}')
        
print('Crawling Completed')
```
